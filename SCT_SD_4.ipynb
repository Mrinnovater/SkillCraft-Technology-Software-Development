```python
from bs4 import BeautifulSoup
import openpyxl

# Read the HTML file
with open('Amazon.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(html_content, 'html.parser')

# Find all the relevant divs
product_sections = soup.find_all('div', class_='a-section')

# Initialize a list to store product details
product_details = []

for section in product_sections:
    product_name = " "
    product_price = " "
    product_reviews = " "

    # Find the product name
    title_div = section.find('div', attrs={'data-cy': 'title-recipe'}, class_='a-section a-spacing-none puis-padding-right-small s-title-instructions-style')
    if title_div:
        product_name = title_div.get_text(strip=True)

    # Find the product price
    price_div = section.find('div', class_='puisg-col puisg-col-4-of-12 puisg-col-4-of-16 puisg-col-4-of-20 puisg-col-4-of-24')
    if price_div:
        product_price = price_div.get_text(strip=True)

    # Find the product reviews
    reviews_div = section.find('div', attrs={'data-cy': 'reviews-block'}, class_='a-section a-spacing-none a-spacing-top-micro')
    if reviews_div:
        product_reviews = reviews_div.get_text(strip=True)

    # Append the product details to the list
    product_details.append((product_name, product_price, product_reviews))

# Create an Excel file and write the product details
workbook = openpyxl.Workbook()
sheet = workbook.active
sheet.title = "Product Details"

# Write the headers
sheet.append(["Product Name", "Product Price", "Product Reviews"])

# Write the product details
for details in product_details:
    sheet.append(details)

# Save the workbook
workbook.save('Amazon_Products.xlsx')
```
